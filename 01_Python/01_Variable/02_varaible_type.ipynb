## 숫자형 (Numeric)

### 숫자 자료형의 종류
* 파이썬에서는 정수, 실수, 복소수 등 숫자를 표현하는 자료형을 모두 숫자 자료형이라고 표현한다.

```python
# 정수
num1 = 1 # int 정수형
num1 = 3.14 

print(type(num1)) # float 실수형
```

### 숫자 자료형의 연산
* 숫자 자료형간에 기본적인 사칙연산과 파이썬에서만 가능한 특수연산을 할 수 있다.

```python
num1 = 10
num2 = 3

print(num1 + num2)
print(num1 - num2)
print(num1 * num2)
print(num1 / num2) # 몫을 구하는 연산
```

```python
print(num1 % num2) # 나머지를 구하는 연산

print(num1 // num2) # 몫만 구하는 연산
```

```python
# 제곱연산
base = 3
print( base ** 2)
```

## 논리형 (Boolean)

### 논리 자료형
* **true** 또는 **false** 값을 가지는 자료형
* 논리 자료형은 문자열이 아니므로 따옴표( "" , '' )로 감싸서는 안되며, 첫 글자를 반드시 대문자로 해야한다.
* 또한 논리 자료형은 자료형 자체로 사용하기도 하지만 비교 연산, 논리 연산의 결과로 논리 자료형이 반환된다.

```python
bol1 = True
bol2 = False

print(type(bol1))
```

### 논리 자료형의 연산

- 비교 연산 : 
    - **==** : 좌항과 우항의 값이 동일한지 비교하여, 같으면 True, 같지 않으면 False를 반환한다.
    - **!=** : 좌항과 우항의 값이 같지 않은지 비교하여, 서로 다르면 True, 같으면 False를 반환한다.
    - **\>,** < : 일반적으로 수학에서 사용하는 부등호와 같이 좌항과 우항의 크고 작음을 비교한다.
    - **\>=, <=** : 같거나 크다 또는 같거나 작다를 비교한다.

```python
a = 123
b = 122

print(a < b)
```

### and 연산과 or 연산

```python
bol1 = False
bol2 = True
```

```python

# and 연산 둘다 True 이어야지 True
print(bol1 & bol2) 
print(bol1 and bol2) 
```

```python
# or 연산 둘중 하나라도 True이면 True
print(bol1 | bol2)
print(bol1 or bol2)
```

```python
print(not bol1) # 불리언값을 반전시킴
```

## 문자형 (String)
### 문자형 선언
* 쌍따옴표("") 또는 홀따옴표('')를 사용하여 문자형 값을 선언할 수 있다.

```python
text1 = '안녕하세요'
text2 = "안녕하세요"

print(text1)
print(text2)
```

* 삼중 따옴표를 사용해 여러줄(멀티라인)으로 출력할 수 있다.

```python
text = """
안녕하세요
반갑습니다
이건 여러줄을 표현하는 문자열(멀티라인)입니다.
"""
print(text)
```

### 문자형 연산

```python
str1 = '아버지가'
str2 = '방에 들어가신다'

print(str1 + ' ' + str2)
```

```python
x = 5 # 정수형
str1 = 'x = ' # 문자형

# print(str1 + x) # 불가능 정수형 + 문자형은 안됨
print(str1 + str(x)) # 정수형을 문자형으로 바꾸어주어야함.
```

f-string: 파이썬 3.6 이상부터 사용가능하며, 문자열 앞에 f를 붙여 사용한다.
- 문자열에 변수의 값을 바로 넣어서 출력이 가능하다.

```python
x = 10

print(f'x = {x}')
```

## 조건문 개요
### 조건문이란?
* 조건에 따라 수행할 코드를 다르게 작성하는 방법으로, 파이썬 키워드를 사용하여 작성한다.
* 조건문 작성에는 조건식이 필요하며, 조건식은 True 혹은 False의 결과를 내는 구문을 의미함.

---
## 조건문 if
### if
* if 키워드 뒤의 조건식의 결과 값이 참(True)이면 실행구문을 실행한다.
* 파이썬에서는 조건식 뒤에 콜론을 붙이고 "반드시" 들여쓰기(공백4칸 또는 tap)로 실행구문을 식별할 수 있도록 작성해야한다.

```python
if 조건식:
    (실행구문)
```

````python
teacher = 'bear'

if teacher == 'tiger':
    # 참일때 실행되는 코드
    print('우리 강사님은 tiger 입니다.')
````

````python
teacher = 'bear'

if teacher == 'tiger':
    # true일때 실행되는 코드
    print('우리 강사님은 tiger 입니다.')
else:
    # false일때 실행되는 코드
    print('우리 강사님은 tiger가 아닙니다.')
````

**elif (if-elif-else)**
* elif 1개 이상 사용 할 수 있다.
* 각각의 경우에 수행할 내용을 구분할 수 있다록 **"반드시"** 들여쓰기를 해주어야 한다

````python
teacher = 'bear'

if teacher == 'tiger':
    # true일때 실행되는 코드
    print('우리 강사님은 tiger 입니다.')
elif teacher == 'bear':
    print('우리 강사님은 bear 입니다.')
else:
    # false일때 실행되는 코드
    print('우리 강사님은 tiger가 아닙니다.')
````

### 삼항연산자

````python
a = 10
b = 10

if a > 10:
    print( "a가 10보다 크다.")
else:
    print( "a가 10보다 크지 않다.")
    
# 삼항연산자
# [참일때 수행될 코드] if [조건식] else [거짓일때 실행될 코드]

print('b가 10보다 크다' if b > 10 else 'b가 10보다 크지 않다')

````

## 조건문 match
### match-case
* 주어진 값을 case 블록의 값과 비교해 일치하는 case만 실행한다.
* input() : 값을 입력받아 변수에 저장

````python
input_food = input('먹고싶은 점심심을 입력해주세요')

match input_food:
    case '짜장면':
        print('짜장면이랑 탕수육을 같이먹고싶어요..')
    case '치킨':
        print('호식이 두마리')
````

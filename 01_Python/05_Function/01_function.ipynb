## 함수 개요

### 함수란?
* 특정 기능을 하기 위한 코드의 집합을 의미한다.
* 함수를 정의해놓으면 해당 기능이 필요한 위치에서 함수를 호출하여 간편하게 사용 가능하다.

### 함수 표현식
* def 키워드를 사용하여 함수를 정의한다.(define)
* 함수 이름 뒤에 소괄호와 콜론을 작성하며 함수에서 실행할 내용은 **"반드시"** 들여쓰기한다.

```python
def basic_function():
    print("Hello world")
    

basic_function()
```

````python
print("안녕하세요")
print("5 + 10 계산을 하겠습니다.")
print(5 + 10)

print("안녕하세요")
print("5 + 10 계산을 하겠습니다.")
print(5 + 10)

print("안녕하세요")
print("5 + 10 계산을 하겠습니다.")
print(5 + 10)
````

````python
def function():
    print("안녕하세요")
    print("5 + 10 계산을 하겠습니다.")
    print(5 + 10)

function()
function()
function()
````

## 매개변수
### 매개변수(파라미터)
* 함수 생성시 함수명 옆에 붙이는 소괄호 안에 매개변수로 전달받을 값의 변수명을 넣을 수 있다.

````python
def cal_sum(x, y):
    print("안녕하세요")
    print(f"{x} + {y} 계산을 하겠습니다.")
    print(x + y)
````

````python
cal_sum(3, 5)
````

* 함수를 정의할때 매개변수의 기본값을 지정 할 수 있다.
    * 함수 호출시 인자를 보내지 않으면 함수 선언부에서 미리 지정한 기본값을 사용하여 함수를 실행한다.

````python
def cal_sum(x = 1, y = 1):
    print("안녕하세요")
    print(f"{x} + {y} 계산을 하겠습니다.")
    print(x + y)

cal_sum(10, 30)
````

### 인자
* 함수 호출 시 함수명에 소괄호를 붙여 호출하며, 이때 소괄호 안에 전달하는 값은 인자라고 한다.

````python
def cal_sum(x = 1, y = 1):
    print("안녕하세요")
    print(f"{x} + {y} 계산을 하겠습니다.")
    print(x + y)
````

````python
# 위치인자
cal_sum(10,30)
cal_sum(30,10)
````

````python
# 키워드 인자

cal_sum(y=5,x=3)
````

## 반환값
### return
* 함수 블럭 내부에서 return 키워드를 만나면 함수의 작동이 중단된다.
* return 뒤에 변수 또는 값이 오면 해당 값을 함수 호출 위치에 반환하며 중단된다.

````python
def cal_sum(x = 1, y = 1):
    result = x+y
    return result
````

````python
print(cal_sum(4,8))
````

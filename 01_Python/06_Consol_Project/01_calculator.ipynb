# 나만의 계산기 만들기 🧮

## 문제 설명

다음 요구사항을 만족해야 합니다:

1. 두 숫자를 더하는 함수 만들기
2. 두 숫자를 빼는 함수 만들기  
3. 두 숫자를 곱하는 함수 만들기
4. 두 숫자를 나누는 함수 만들기
5. 모든 함수를 사용해서 계산기 만들기
- 계산기 시작화면
    ```text
        ========================================
        🧮 나만의 계산기에 오신 것을 환영합니다! 🧮

        사용 가능한 연산:
        1. 덧셈 (+)
        2. 뺄셈 (-)
        3. 곱셈 (×)
        4. 나눗셈 (÷)

        ==============================
    ```
    - 계산기는 항상 계산이 끝난뒤 반복할지 안할지를 물어보고
    - 아니면 종료됨

    - 실행예시
    ```text
        ========================================
        🧮 나만의 계산기에 오신 것을 환영합니다! 🧮

        사용 가능한 연산:
        1. 덧셈 (+)
        2. 뺄셈 (-)
        3. 곱셈 (×)
        4. 나눗셈 (÷)

        ==============================
        1.0 + 3.0 = 4.0
        계산기를 사용해주셔서 감사합니다!
    ```

## 힌트
- def 키워드로 함수 정의하기
- 함수 안에서 return으로 결과 반환하기
- 함수를 호출할 때는 함수명() 사용하기
- 매개변수는 괄호 안에 넣기
- while문으로 반복, break는 탈출!

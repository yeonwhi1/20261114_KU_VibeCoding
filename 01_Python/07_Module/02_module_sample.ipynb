다음은 참고용 예시입니다. 직접 만들어보세요!

```python
def square(x):
    """제곱을 구하는 함수"""
    return x * x

def cube(x):
    """세제곱을 구하는 함수"""
    return x * x * x

def power(x, n):
    """x의 n제곱을 구하는 함수"""
    result = 1
    for i in range(n):
        result = result * x
    return result


### mymodule 사용 예시

```python
# mymodule을 만들었다면 이렇게 사용할 수 있습니다


```
